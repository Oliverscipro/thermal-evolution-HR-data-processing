```python
with h5py.File(file, "r") as f:
    x = f["FullSpectra/MassAxis"][:]
    y = f["FullSpectra/SumSpectrum"][:]
    peak_table = f["PeakData/PeakTable"][:]

# Filter data to mass range 40-200
mask = (x >= 40) & (x <= 200)
x_filtered = x[mask]
y_filtered = y[mask]

# Create the plot with styling similar to your figure
plt.figure(figsize=(12, 8))

# Plot the spectrum as vertical lines (stem plot style)
plt.vlines(x_filtered, 0, y_filtered, colors='black', linewidth=0.8, alpha=0.7)

# Add baseline
plt.axhline(y=0, color='black', linewidth=0.5)

# Filter and plot detected peaks if available
if peak_table.dtype.names:
    if "Mass" in peak_table.dtype.names and "Intensity" in peak_table.dtype.names:
        # Filter peaks to the same mass range
        peak_mask = (peak_table["Mass"] >= 40) & (peak_table["Mass"] <= 200)
        filtered_peaks_mass = peak_table["Mass"][peak_mask]
        filtered_peaks_intensity = peak_table["Intensity"][peak_mask]
        
        # Plot peak markers
        plt.scatter(filtered_peaks_mass, filtered_peaks_intensity,
                   color="red", marker="o", s=30, alpha=0.8, 
                   label="Detected Peaks", zorder=5)
        
        # Optionally add peak labels for significant peaks
        # You can adjust the threshold based on your data
        intensity_threshold = np.percentile(filtered_peaks_intensity, 90)
        for mass, intensity in zip(filtered_peaks_mass, filtered_peaks_intensity):
            if intensity > intensity_threshold:
                plt.annotate(f'{mass:.3f}', 
                           xy=(mass, intensity), 
                           xytext=(5, 5), 
                           textcoords='offset points',
                           fontsize=8, 
                           color='blue',
                           ha='left')

# Styling to match your figure
plt.xlabel("m/z", fontsize=12, fontweight='bold')
plt.ylabel("Intensity [a.u.]", fontsize=12, fontweight='bold')
plt.title("Mass Spectrum (m/z 40-200)", fontsize=14, fontweight='bold')

# Set axis limits
plt.xlim(40, 200)
plt.ylim(0, np.max(y_filtered) * 1.1)

# Grid for better readability
plt.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)

# Improve tick marks
plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(np.arange(40, 201, 20))  # Major ticks every 20 m/z units

# Add minor ticks
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', length=3)
plt.tick_params(axis='y', which='minor', length=3)

# Legend
if peak_table.dtype.names and "Mass" in peak_table.dtype.names:
    plt.legend(loc='upper right', fontsize=10)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

# Optional: Print some statistics
print(f"Mass range: {x_filtered.min():.3f} - {x_filtered.max():.3f}")
print(f"Number of data points in range: {len(x_filtered)}")
if peak_table.dtype.names and "Mass" in peak_table.dtype.names:
    peak_mask = (peak_table["Mass"] >= 40) & (peak_table["Mass"] <= 200)
    print(f"Number of detected peaks in range: {np.sum(peak_mask)}")
```

```python
with h5py.File(file, "r") as f:
    x = f["FullSpectra/MassAxis"][:]
    y = f["FullSpectra/SumSpectrum"][:]
    peak_table = f["PeakData/PeakTable"][:]

# Filter data to mass range 40-200
mask = (x >= 40) & (x <= 200)
x_filtered = x[mask]
y_filtered = y[mask]

# Create the plot with styling similar to your figure
plt.figure(figsize=(12, 8))

# Plot the spectrum as vertical lines (stem plot style)
plt.vlines(x_filtered, 0, y_filtered, colors='black', linewidth=0.8, alpha=0.7)

# Add baseline
plt.axhline(y=0, color='black', linewidth=0.5)

# Filter and annotate peaks based on spectrum intensity > 1e9
if peak_table.dtype.names:
    print("PeakTable dtype:", peak_table.dtype)
    print("Available peak table columns:", peak_table.dtype.names)
    
    # Extract peak masses and labels from the peak table
    peak_masses = peak_table['mass']
    peak_labels = peak_table['label']  # This contains the peak names/identifiers
    
    # Filter peaks to mass range 40-200
    mass_range_mask = (peak_masses >= 40) & (peak_masses <= 200)
    filtered_peak_masses = peak_masses[mass_range_mask]
    filtered_peak_labels = peak_labels[mass_range_mask]
    
    print(f"Found {len(filtered_peak_masses)} peaks in m/z range 40-200")
    
    # For each peak in the table, find corresponding intensity in the spectrum
    significant_peaks = []
    for peak_mass, peak_label in zip(filtered_peak_masses, filtered_peak_labels):
        # Find the closest point in the spectrum to this peak mass
        closest_idx = np.argmin(np.abs(x_filtered - peak_mass))
        spectrum_intensity = y_filtered[closest_idx]
        
        # Only include if intensity > 1e9
        if spectrum_intensity > 3e5:
            # Decode bytes to string if necessary
            label_str = peak_label.decode('utf-8') if isinstance(peak_label, bytes) else str(peak_label)
            significant_peaks.append((peak_mass, spectrum_intensity, label_str))
    
    print(f"Found {len(significant_peaks)} peaks with spectrum intensity > 1e9")
    
    if significant_peaks:
        # Extract masses, intensities, and labels for plotting
        sig_masses = [peak[0] for peak in significant_peaks]
        sig_intensities = [peak[1] for peak in significant_peaks]
        sig_labels = [peak[2] for peak in significant_peaks]
        
        # Plot peak markers for significant peaks
        plt.scatter(sig_masses, sig_intensities,
                   color="red", marker="o", s=60, alpha=0.9, 
                   label=f"Peaks (Intensity > 1e9)", zorder=5, 
                   edgecolors='darkred', linewidth=2)
        
        # Add peak labels using the label field from peak table
        for i, (mass, intensity, label) in enumerate(significant_peaks):
            # Alternate label positions to avoid overlap
            y_offset = 20 if i % 2 == 0 else -25
            x_offset = (-1)**(i//2) * 8  # Slight horizontal offset for very close peaks
            
            # Use the peak label from the table
            plt.annotate(label, 
                       xy=(mass, intensity), 
                       xytext=(x_offset, y_offset), 
                       textcoords='offset points',
                       fontsize=10, 
                       color='darkblue',
                       ha='center',
                       weight='bold',
                       bbox=dict(boxstyle='round,pad=0.4', 
                               facecolor='lightyellow', 
                               alpha=0.9,
                               edgecolor='darkblue',
                               linewidth=1),
                       arrowprops=dict(arrowstyle='->', 
                                     color='darkblue', 
                                     alpha=0.8,
                                     lw=1.5))
        
        # Print peak information with labels
        print("\nSignificant peaks (spectrum intensity > 1e9):")
        print("Label\t\tm/z\t\tIntensity")
        print("-" * 45)
        for mass, intensity, label in sorted(significant_peaks, key=lambda x: x[1], reverse=True):
            print(f"{label}\t\t{mass:.4f}\t{intensity:.2e}")
    else:
        print("No peaks found with spectrum intensity > 1e9")
else:
    print("No peak table data available")

# Styling to match your figure
plt.xlabel("m/z", fontsize=12, fontweight='bold')
plt.ylabel("Intensity [a.u.]", fontsize=12, fontweight='bold')
plt.title("Mass Spectrum (m/z 40-200)", fontsize=14, fontweight='bold')

# Set axis limits
plt.xlim(40, 200)
plt.ylim(0, np.max(y_filtered) * 1.1)

# Grid for better readability
plt.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)

# Improve tick marks
plt.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(np.arange(40, 201, 20))  # Major ticks every 20 m/z units

# Add minor ticks
plt.minorticks_on()
plt.tick_params(axis='x', which='minor', length=3)
plt.tick_params(axis='y', which='minor', length=3)

# Legend
if peak_table.dtype.names and len([mass for mass in peak_table['mass'] if 40 <= mass <= 200]) > 0:
    plt.legend(loc='upper right', fontsize=10)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

# Optional: Print some statistics
print(f"Mass range: {x_filtered.min():.3f} - {x_filtered.max():.3f}")
print(f"Number of data points in range: {len(x_filtered)}")
print(f"Maximum intensity in range: {np.max(y_filtered):.2e}")
if peak_table.dtype.names:
    peak_masses = peak_table['mass']
    mass_range_mask = (peak_masses >= 40) & (peak_masses <= 200)
    peaks_in_range = np.sum(mass_range_mask)
    print(f"Total peaks in mass range 40-200: {peaks_in_range}")
```

```python
file = r"C:\Users\barrio_o\Documents\Oliver\python\h5_data\processed_2024_12_10_glucose_calibration\20241210_130447_glucose_0.15umol_p.h5"

with h5py.File(file, "r") as f:
    # 1. Peak list (m/z values)
    peak_table = f["PeakData/PeakTable"][:]
    masses = peak_table["mass"]

    # 2. Peak intensities
    peak_data = f["PeakData/PeakData"][:, :, 0, :]  # (2082, 2, 1, 1611)
    peak_data = peak_data.reshape(-1, peak_data.shape[2])     # -> (2082, 1611)

    # 3. Times (choose first column if shape is (2082,2))
    buf_times = np.array(f["TimingData/BufTimes"][:].flatten())
    print("BufTimes shape:", buf_times.shape)

    if buf_times.ndim == 2 and buf_times.shape[1] == 2:
        buf_times = buf_times[:, 0]   # take the first column

# 4. Build DataFrame
df = pd.DataFrame(
    peak_data,
    columns=[f"{m:.4f}" for m in masses]
)

# Insert time column
df.insert(0, "time", buf_times[:len(df)])

# Save to CSV
df.to_csv("glucose_peaks_timeseries.csv", index=False)


print(df.head())
```
